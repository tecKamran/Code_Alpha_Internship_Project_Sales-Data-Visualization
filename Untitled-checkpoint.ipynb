```python

#  IMPORT LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  STEP 1: LOAD THE DATA
data = pd.read_csv(r"C:\Users\ADMIN\Documents\Code_Alpha_project2\sales_data.csv")

#  STEP 2: CHECK DATA
print("First 5 Rows:")
print(data.head())

print("\nDataset Info:")
print(data.info())

#  STEP 3: CLEAN DATA (CONVERT DATE)
data["Date"] = pd.to_datetime(data["Date"])

#  STEP 4: BAR CHART  SALES BY PRODUCT
plt.figure(figsize=(6,4))
sns.barplot(x="Product", y="Sales", data=data)
plt.title("Total Sales by Product")
plt.show()

#  STEP 5: LINE CHART  SALES TREND OVER TIME
plt.figure(figsize=(8,4))
plt.plot(data["Date"], data["Sales"])
plt.xticks(rotation=45)
plt.title("Sales Trend Over Time")
plt.ylabel("Sales")
plt.show()

#  STEP 6: SCATTER PLOT  SALES vs PROFIT
plt.figure(figsize=(6,4))
sns.scatterplot(x="Sales", y="Profit", data=data)
plt.title("Relationship Between Sales & Profit")
plt.show()

#  STEP 7: BAR CHART  SALES BY REGION
plt.figure(figsize=(6,4))
sns.barplot(x="Region", y="Sales", data=data)
plt.title("Sales by Region")
plt.show()

#  STEP 8: HEATMAP  CORRELATION
plt.figure(figsize=(6,4))
corr = data.select_dtypes(include=["int64", "float64"]).corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()


#  STEP 9: SMALL DASHBOARD (MULTIPLE CHARTS)
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.barplot(x="Product", y="Profit", data=data)
plt.title("Profit by Product")

plt.subplot(1,2,2)
sns.barplot(x="Region", y="Profit", data=data)
plt.title("Profit by Region")

plt.tight_layout()
plt.show()

print("\n Project Completed Successfully!")
```

```python
product_sales = data.groupby("Product")["Sales"].sum()

plt.figure(figsize=(6,6))
plt.pie(product_sales, labels=product_sales.index, autopct='%1.1f%%')
plt.title("Sales Contribution by Product")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Region", y="Sales", data=data)
plt.title("Sales Distribution by Region")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data["Profit"], kde=True)
plt.title("Profit Distribution")
plt.show()
```
